```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
d = [np.sin(np.linspace(0, 2*i * np.pi , 10000) ) / (2*i) for i in range(1, 32)]
```

```python

plt.figure(figsize=(30,10))
sns.lineplot(data=d)
```
